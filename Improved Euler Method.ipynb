## Improved Euler Method

```python
import numpy as np
import matplotlib.pyplot as plt

def dydx(x, y):
    return np.sin(x) + np.cos(x)

x = 0
y = 1
h = 0.001

x_plot = []
y_plot = []

while x <= 15:
    x_plot.append(x)
    y_plot.append(y)
    y = y + h * dydx(x + (h / 2), y + h * dydx(x, y) / 2)
    x = x + h
    
plt.plot(x_plot, y_plot, label='Solution Plot')
plt.xlabel('x', fontdict={'family':'serif', 'color':'black', 'fontsize':15})
plt.ylabel('y', fontdict={'family':'serif', 'color':'black', 'fontsize':15})
plt.legend(shadow=True, prop={'family':'serif', 'size':10})
plt.grid()
plt.show()
```
