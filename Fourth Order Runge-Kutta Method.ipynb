## Fourth Order Runge-Kutta Method

```python
import numpy as np
import matplotlib.pyplot as plt

def dydx(x,y):
    return x + y

x = 1
y = 1
h = 0.001

x_plot = []
y_plot = []

while x < 5:
    x_plot.append(x)
    y_plot.append(y)
    k1 = h * dydx(x, y)
    k2 = h * dydx(x + (h / 2), y + (k1 / 2))
    k3 = h * dydx(x + (h / 2), y + (k2 / 2))
    k4 = h * dydx(x + h, y + k3)
    y = y + ((k1 + 2 * k2 + 2 * k3 + k4) / 6)
    x = x + h
    
plt.plot(x_plot, y_plot, label='Solution Plot')
plt.xlabel('x', fontdict={'family':'serif', 'color':'black', 'fontsize':15})
plt.ylabel('y', fontdict={'family':'serif', 'color':'black', 'fontsize':15})
plt.legend(shadow=True, prop={'family':'serif', 'size':10})
plt.grid()
plt.show()
```
