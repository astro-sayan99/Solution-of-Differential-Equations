## Simple Euler Method

```python
import numpy as np
import matplotlib.pyplot as plt

def dxdt(t, a=2):
    return a*np.sin(t)

t = 0
x = 0
h = 0.001

t_plot = []
x_plot = []

while t <= 4*np.pi:
    t_plot.append(t)
    x_plot.append(x)
    x = x + h*dxdt(t, a=2)
    t = t + h
    
plt.plot(t_plot, x_plot, label='Solution Plot')
plt.xlabel('t', fontdict={'family':'serif', 'color':'black', 'fontsize':15})
plt.ylabel('x', fontdict={'family':'serif', 'color':'black', 'fontsize':15})
plt.legend(shadow=True, prop={'family':'serif', 'size':10})
plt.grid()
plt.show()
```
