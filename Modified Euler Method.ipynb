## Modified Euler Method

```python
import numpy as np
import matplotlib.pyplot as plt

def dydx(x, y):
    return x + y

x = 0
y = 1
h = 0.001

x_plot = []
y_plot = []

while x <= 5:
    x_plot.append(x)
    y_plot.append(y)
    y = y + (h / 2) * (dydx(x, y) + dydx(x + h, y + h * dydx(x, y)))
    x = x + h
    
plt.plot(x_plot, y_plot, label='Solution Plot')
plt.xlabel('x', fontdict={'family':'serif', 'color':'black', 'fontsize':15})
plt.ylabel('y', fontdict={'family':'serif', 'color':'black', 'fontsize':15})
plt.legend(shadow=True, prop={'family':'serif', 'size':10})
plt.grid()
plt.show()
```
